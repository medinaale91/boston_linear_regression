```python
# We start by loading the data and investigating it.
boston = load_boston()
boston.keys()
```

```python
boston
# Dictionary-like object, the interesting attributes are: ‘data’, the data to learn, 
# ‘target’, the regression targets, and ‘DESCR’, the full description of the dataset.
```

```python
print(boston.data.shape)
```

```python
df = pd.DataFrame(boston.data)
df.head()
```

```python
# We add the column names which are located within the dictionary key "feature names"
df.columns = boston.feature_names
df.head()
```

```python
print(boston.target.shape)
```

```python
# We add "Target" column into the dataframe.
df["target"] = boston.target
df.head()
```

```python
# checking the data for null values
df.isnull().sum()
```

```python
# We want to find out the correlation between the columns.
correlation = df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'target']].corr()
correlation
```

```python
#another option showing only the correlation against the column "target": 

df.corr()['target']
# We can see that the columns that are most correlationated to target are "LSTAT" (-0.7376) 
# & "RM" (0.6953) as they are the closest to 1 or -1.
```

```python
##Comparing "LSTAT" against "TARGET" with statsmodels
```

```python
mod = smf.OLS(df["LSTAT"], df["target"])
```

```python
res = mod.fit()
```

```python
res.summary()
```

```python
# Before splitting the data to train-test dataset, we would need to split the dataset into 
# two: target value and predictor values. We will call the target value "Y" and predictor 
# values "X".
X = df.drop("target",axis=1)
Y = df["target"]
```

```python
# We split the dataset into train and test 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
# If we also check the shape of each variable, we can find our train and test datasets
# with the proportion of 80.00% for train data and 20.00% for test data.
```

```python

lm = LinearRegression(fit_intercept=True)
lm.fit(X_train,Y_train)

Y_pred = lm.predict(X_test)

plt.scatter(Y_test, Y_pred)
plt.xlabel("Prices")
plt.ylabel("Predicted prices")
plt.title("Prices vs Predicted prices")
```

```python
r2 = r2_score(y_pred=y_pred, y_true=Y)
r2
```

```python
rmse = mean_squared_error(y_pred=y_pred, y_true=Y)**.5
rmse
```
